```python
def getUrlIMDBtoprated50movies(url_link):
    imdb_page=requests.get(url_link)
    imdb_content=BeautifulSoup(imdb_page.content)
    #imdb_tables = imdb_content.find_all("table")
    imdb_title_col = imdb_content.find_all("td", class_="titleColumn")
    rank_list=[]
    name_list=[]
    year_list=[]
    for i in imdb_title_col:
        #print(i.text.split("\n"))
        movie_list=i.text.split("\n")
        rank_list.append(movie_list[1].replace('.','').strip())
        name_list.append(movie_list[2].strip())
        year_list.append(movie_list[3].replace('(','').replace(")",""))
    imdb_rating_col = imdb_content.find_all("td", class_="ratingColumn imdbRating")
    rating_list=[]
    for i in imdb_rating_col:
        rating_list.append(i.text.split("\n")[1])
    rank_list_top50=rank_list[0:50]
    name_list_top50=name_list[0:50]
    year_list_top50=year_list[0:50]
    rating_list_top50=rating_list[0:50]
    movie_dict={"rank":rank_list_top50,"movie_name":name_list_top50,"year_of_release":year_list_top50,"imdb_rating":rating_list_top50}
    df_imdb50=pd.DataFrame(movie_dict)
    return df_imdb50
```

```python
def getPresidentDetails(urlLink):
    presidentPage=requests.get(urlLink)
    presidentContent=BeautifulSoup(presidentPage.content)
    presidentDivCol = presidentContent.find_all("div", class_="presidentListing")
    pNameList=[]
    pTOFList=[]
    for i in presidentDivCol:
        pList=i.text.split('\n')
        pName=pList[1][0:pList[1].find('(')].strip()
        pTOF=pList[2][pList[2].find(":")+1:].strip()
        pNameList.append(pName)
        pTOFList.append(pTOF)
    president_dict={"Name":pNameList,"TermOfOffice":pTOFList}
    df_president=pd.DataFrame(president_dict)
    return df_president
```

```python
def topOdiPlayers(url_link,ptype):
    odi_men_player_page=requests.get(url_link)
    odi_men_player_content=BeautifulSoup(odi_men_player_page.content)
    odi_table_col_1 = odi_men_player_content.find_all("div", class_="rankings-block__container")
    if ptype == 'batsmen':
        odi_table_col_p=odi_table_col_1[0]
    elif ptype == 'bowler':
        odi_table_col_p=odi_table_col_1[1]
    elif ptype == 'all-rounder':
        odi_table_col_p=odi_table_col_1[2]
    else:
        print("Type of player is incorrect.")
    nl=[]
    cl=[]
    rl=[]
    for i in odi_table_col_p.find_all("div", class_="rankings-block__top-player"):
        fl=i.text.split('\n')
        name= fl[13]
        country=fl[16]
        rating=fl[17].strip()
        nl.append(name)
        cl.append(country)
        rl.append(rating)
    for i in odi_table_col_p.find_all("tr", class_="table-body"):
        fl=i.text.split('\n')
        #print(fl)
        if fl[13] != '':
            name= fl[13]
            country=fl[17]
            rating=fl[19]
        else:
            name= fl[16]
            country=fl[20]
            rating=fl[22]
        nl.append(name)
        cl.append(country)
        rl.append(rating)
    odi_table_col_bat_dict={"name":nl,"country":cl,"rating":rl}
    df_bat=pd.DataFrame(odi_table_col_bat_dict)
    return df_bat 
```

```python
df_bat=topOdiPlayers('https://www.icc-cricket.com/rankings/mens/player-rankings/odi','batsmen')
df_bat
```
