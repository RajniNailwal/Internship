### ANSWER 1

```python
driver=webdriver.Chrome("chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
try:
    videoTable=driver.find_element(By.XPATH,"/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody")
except NoSuchElementException:
    print("NoSuchElementException occurred")
except:
    print("An exception occurred")
```

```python
videoTableRows=videoTable.find_elements(By.XPATH,"./tr")
```

```python
listRank=[]
listName=[]
listArtist=[]
listUploadDate=[]
listViews=[]
```

```python
for i in videoTableRows:
    rowDetails=i.find_elements(By.XPATH,"./td")
    listRank.append(rowDetails[0].text.replace(".",""))
    listName.append(rowDetails[1].text.split("[")[0].replace('"',''))
    listArtist.append(rowDetails[2].text)
    listViews.append(rowDetails[3].text)
    listUploadDate.append(rowDetails[4].text)
```

```python
dict_wiki={"Rank":listRank,"Name":listName,"Artist":listArtist,"UploadDate":listUploadDate,"Views":listViews}
df_wiki=pd.DataFrame(dict_wiki)
df_wiki
```
